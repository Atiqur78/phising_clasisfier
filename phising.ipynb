```python
plt.figure(figsize=(15,50), facecolor='white')
plotnumber = 1

for column in df.drop(['Result'], axis =1):
    ax = plt.subplot(12,3,plotnumber)
    sns.countplot(x=df[column],hue= df['Result'])
    plt.xlabel(column, fontsize = 10)
    plotnumber+=1
plt.show()
```
